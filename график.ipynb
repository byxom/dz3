```python
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as pyplot
import matplotlib.patheffects as path_effects
from matplotlib.pyplot import figure
#импортирую бибилиотеки

figure(num=None, figsize=(10, 5), dpi=80, facecolor='w', edgecolor='k')
#фикисрую размер графика и все такое 
ax = pyplot.subplot()
np.random.seed(19680801)
#фиксирую генератор случайных чисел
x=np.linspace(1,5000,100)
y1= [np.random.random_sample() for i in x]
y2= [i-np.random.random_sample()**4*i for i in y1]
y3= np.random.uniform(0, 0.3, 100)
#генерирую оси x и y 
ax.plot(x, y2, color='moccasin')
ax.plot(x, y1,color='#f95b2e')
ax.plot(x, y3,color='silver')
#рисую график
clr='#ef5d26'

an=ax.annotate("случайная точка",
            color= clr,
            fontsize= 14,
            xy=(500, 0.5),
            xytext=(500, 0.8),
            arrowprops=dict(arrowstyle='fancy,head_length=1.5,head_width=1,tail_width=0.1',
                                color= 'red',
                                shrinkA=10,
                                shrinkB=20,
                                connectionstyle="arc3",
                                linewidth=3,
                                relpos=(0.1,0.5),path_effects=[path_effects.withSimplePatchShadow(offset=(4, -2.3)),
                     path_effects.withStroke(linewidth=6,foreground="w")]))
#добавил аннотацию к графику -текст и стрелку с тенью, обводкой и смещением
an1=ax.annotate("еще одна случайная точка",
            color= clr,
            fontsize= 14,
            xy=(3000, 0.5),
            xytext=(3000, 0.8),
            arrowprops=dict(arrowstyle='fancy,head_length=1.5,head_width=1,tail_width=0.1',
                                color= 'red',
                                shrinkA=10,
                                shrinkB=20,
                                connectionstyle="arc3",
                                linewidth=3,
                                relpos=(0.1,0.5),path_effects=[path_effects.withSimplePatchShadow(offset=(4, -2.3)),
                     path_effects.withStroke(linewidth=6,foreground="w")]))
#еще одну такую же аннотацию 
an.set_path_effects([path_effects.withSimplePatchShadow(offset=(1.9, -2.3)),
                     path_effects.withStroke(linewidth=3,foreground="w")])
#добавил тень и обводку на текст
an1.set_path_effects([path_effects.withSimplePatchShadow(offset=(1.9, -2.3)),
                     path_effects.withStroke(linewidth=3,foreground="w")])
#то же самое для второго текста
pyplot.title("График")
#озаглавил график 
```
